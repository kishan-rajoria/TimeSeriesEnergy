```python
X['Intercept'] = 1
```

```python
vif_data = pd.DataFrame()
vif_data['Feature'] = X.columns
vif_data['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
```

```python
X = X.drop(columns='Intercept')
```

```python
print(vif_data)
```

```python
high_vif_features = vif_data[vif_data['VIF'] > 10]['Feature'].tolist()
X_filtered = X.drop(columns=high_vif_features)
```

```python
df_filtered = df.drop(columns=high_vif_features)
```

```python
df_train = df_filtered[['date_time', 'substation_mtr_c1_p_kw']].rename(columns={'date_time': 'ds', 'substation_mtr_c1_p_kw': 'y'})
```

```python
m = NeuralProphet()
```

```python
for feature in X_filtered.columns:
    m = m.add_future_regressor(name=feature)
```

```python
df_train
```

```python
metrics = m.fit(df_train, freq='T')
```

```python
df_train['ds'] = pd.to_datetime(df_train['ds'], errors='coerce', utc=True)
```

```python
future = m.make_future_dataframe(df_train, periods=7*24*60)
```

```python
forecast = m.predict(future)
```

```python
fig = m.plot(forecast)
```

```python
fig.update_layout(
    title="Forecast vs Historical Data",
    xaxis_title="Date",
    yaxis_title="Substation Power (mw)",
    legend_title="Legend"
)
fig.show()
```
